```python
count_class_0,count_class_1=df.type.value_counts()
data_class_0=df[df["type"]==0]
data_class_1=df[df["type"]==1]
```

```python
count_class_0,count_class_1

data_class_under=data_class_0.sample(count_class_1)
data_under=pd.concat([data_class_under,data_class_1],axis=0)

df
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def pre(desc):
  d=nlp(desc)
  filtered_tokens=[]
  for token in d:
    if token.is_stop or token.is_punct:
      continue
    filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)
```

```python
data_under["New_SMS"]=data_under.SMS.apply(pre)
```

```python
df
```

```python
data_under=data_under.drop("SMS",axis=1)
data_under
```

```python
from  sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test = train_test_split(data_under.New_SMS,data_under.type,test_size=0.2,random_state=1,stratify=data_under.type    )
```

```python
print("\nMissing values in X_train:")
print(X_train.isnull().sum())
print("\nMissing values in Y_train:")
print(Y_train.isnull().sum())
```

```python
X_train = X_train.fillna('')  # Replace NaN values with empty strings
Y_train = Y_train.fillna(0)   # Replace NaN values with 0 (or appropriate imputation)
```

```python
print("\nMissing values in X_train:")
print(X_train.isnull().sum())
print("\nMissing values in Y_train:")
print(Y_train.isnull().sum())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Model=Pipeline(
[
    ('vectorizer_tfidf',TfidfVectorizer()),
    ('Multi NB',MultinomialNB())
])
Model.fit(X_train,Y_train)
y_pred=Model.predict(x_test)
print(classification_report(y_test,y_pred))
```
